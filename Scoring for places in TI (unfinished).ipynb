`Arrow Up/Arrow Down` : Navigate between cells.
`A`: Insert a new cell above the current cell.
`B`: Insert a new cell below the current cell.
`C`: Copy the selected cell.
`V`: Paste the copied cell below the current cell.
`D`: Delete the selected cell.

```python
TI_2011_pl = pd.read_csv('TI_ScrappedData/TI_2011/PrizePoolFinalStandings.csv')
TI_2012_pl = pd.read_csv('TI_ScrappedData/TI_2012/PrizePoolFinalStandings.csv')
TI_2013_pl = pd.read_csv('TI_ScrappedData/TI_2013/PrizePoolFinalStandings.csv')
TI_2014_pl = pd.read_csv('TI_ScrappedData/TI_2014/PrizePoolFinalStandings.csv')
TI_2015_pl = pd.read_csv('TI_ScrappedData/TI_2015/PrizePoolFinalStandings.csv')
TI_2016_pl = pd.read_csv('TI_ScrappedData/TI_2016/PrizePoolFinalStandings.csv')

TI_2017_pl = pd.read_csv('TI_ScrappedData/TI_2017/PrizePoolFinalStandings.csv')
TI_2018_pl = pd.read_csv('TI_ScrappedData/TI_2018/PrizePoolFinalStandings.csv')
TI_2019_pl = pd.read_csv('TI_ScrappedData/TI_2019/PrizePoolFinalStandings.csv')
TI_2021_pl = pd.read_csv('TI_ScrappedData/TI_2021/PrizePoolFinalStandings.csv')

TI_pl_list = [TI_2011_pl,TI_2012_pl,TI_2013_pl,TI_2014_pl,TI_2015_pl,
              TI_2016_pl,TI_2017_pl,TI_2018_pl,TI_2019_pl,TI_2021_pl]
```

```python
#A bit did it for usable preference 

def rename_TI_pl(TI_pl):
    TI_pl.columns = ['remove','Place','Price','PricedPoolPercnet','TeamName']
    TI_pl.drop(columns=['remove'], inplace=True)
    #TI_pl.rename_axis('ID', axis = 'columns')
    return TI_pl
```

```python
for TI_pl in TI_pl_list:
    rename_TI_pl(TI_pl)
```

```python
TI = pd.concat(TI_pl_list, axis=0)
#TI['TeamName'].drop_duplicates().sort_values()
```

```python
#TI['Place'].sort_values(ascending = False)

my_TI_scores = TI
my_TI_scores.sort_values('Place', ascending=False)
my_TI_scores.Place.unique()
```

```python
warnings.simplefilter('ignore')

def pts_metric(my_TI_scores):
    my_TI_scores['my_pts'] = ''
    my_TI_scores.loc[my_TI_scores['Place']=='1st','my_pts'] = 10
    my_TI_scores.loc[my_TI_scores['Place']=='2nd','my_pts'] = 9
    my_TI_scores.loc[my_TI_scores['Place']=='3rd','my_pts'] = 8
    my_TI_scores.loc[my_TI_scores['Place']=='4th','my_pts'] = 7
    my_TI_scores.loc[my_TI_scores['Place']=='5th-6th','my_pts'] = 6
    my_TI_scores.loc[my_TI_scores['Place']=='7th-8th','my_pts'] = 5

    my_TI_scores.loc[my_TI_scores['Place']=='9th-12th','my_pts'] = 4
    my_TI_scores.loc[my_TI_scores['Place']=='9th-10th','my_pts'] = 4
    my_TI_scores.loc[my_TI_scores['Place']=='11th-12th','my_pts'] = 4

    my_TI_scores.loc[my_TI_scores['Place']=='13th-16th','my_pts'] = 3
    my_TI_scores.loc[my_TI_scores['Place']=='13th-14th','my_pts'] = 3
    my_TI_scores.loc[my_TI_scores['Place']=='15th-16th','my_pts'] = 3

    my_TI_scores.loc[my_TI_scores['Place']=='17th','my_pts'] = 2
    my_TI_scores.loc[my_TI_scores['Place']=='17th-18th','my_pts'] = 2
    my_TI_scores.loc[my_TI_scores['Place']=='18th-19th','my_pts'] = 2
    return my_TI_scores

#my_TI_scores.groupby('TeamName').my_pts.agg([sum, np.mean, np.median]).sort_values(by=['mean','sum','median'], ascending=False)
```
